<h3> Ultimate Music Database Web Scraper </h3>
- This web scraper is designed to scrape Billboard Hot 100 Data for certain date ranges from the following website:

http://www.umdmusic.com/default.asp?Lang=English&Chart=B

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.select import Select
import pandas as pd
import numpy as np
import time
```

```python
options = webdriver.ChromeOptions() 
options.add_argument(r"download.default_directory=C:\Users\nicholas\Desktop\Python\WebScraper\Music_Data_UpWork\data")
```

```python
driver = webdriver.Chrome(r'C:\Users\nicholas\Downloads\chromedriver_win32(1)\chromedriver.exe', chrome_options=options)  # Optional argument, if not specified will search path.
driver.get("http://www.umdmusic.com/default.asp?Lang=English&Chart=D&ChDate=20181222&ChMode=N")

for j in range(1,52):
    for i in range(3,103):
        row1 = driver.find_element_by_xpath("//table[4]/tbody[1]/tr[" + str(i) + "]/td[1]").text
        row2 = driver.find_element_by_xpath("//table[4]/tbody[1]/tr[" + str(i) + "]/td[2]").text
        row3 = driver.find_element_by_xpath("//table[4]/tbody[1]/tr[" + str(i) + "]/td[3]").text
        row4 = driver.find_element_by_xpath("//table[4]/tbody[1]/tr[" + str(i) + "]/td[4]").text
        row5 = driver.find_element_by_xpath("//table[4]/tbody[1]/tr[" + str(i) + "]/td[5]").text
        row6 = driver.find_element_by_xpath("//table[4]/tbody[1]/tr[" + str(i) + "]/td[6]").text
        row7 = driver.find_element_by_xpath("//table[4]/tbody[1]/tr[" + str(i) + "]/td[7]").text
        row8 = driver.find_element_by_xpath("//table[4]/tbody[1]/tr[" + str(i) + "]/td[8]").text
        row9 = driver.find_element_by_xpath("//table[4]/tbody[1]/tr[" + str(i) + "]/td[9]").text

        title = row5.partition("\n")[0]
        artist = row5.partition("\n")[2]

        date = driver.find_element_by_xpath("//td[2]/b[1]").text

        df = pd.DataFrame(np.array([[row1, row2, row3, row4, title, artist, row6, row7, row8, row9, date]]), columns=['ThisWeekPosition','LastWeekPosition','PeakPosition','TotalWeeks','Title', 'Artist','EntryDate','EntryPosition','PeakPosition','TotalWeeks', 'Week']).append(df, sort=False)


    df.sort_values(by=['Week','ThisWeekPosition'])
    elem = driver.find_element_by_xpath("//table[3]/tbody[1]/tr[1]/td[1]/a[1]/b[1]").click()
```
