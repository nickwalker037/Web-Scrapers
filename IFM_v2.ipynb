```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.select import Select
import pandas as pd
import numpy as np
import time
```

```python
options = webdriver.ChromeOptions() 
options.add_argument(r"download.default_directory=C:\Users\nicholas\Desktop\Python\WebScraper\Medical_Data_UpWork\data")

```

```python
driver = webdriver.Chrome(r'C:\Users\nicholas\Downloads\chromedriver_win32(1)\chromedriver.exe', chrome_options=options)  # Optional argument, if not specified will search path.

state = "WI"
driver.get("https://www.ifm.org/find-a-practitioner/?country=US&city=&province=&state_us="+str(state)+"&state_ca=&postal_code=&pos=&advanced_search=&ifm_certified=&practitioner-first-name=&practitioner-last-name=&insurance=&medicare=&online=&phone=&primary-degree=&languages=")



for j in range(1,11):
    for i in range(1,17):
        person_id = "//div[" + str(i) + "]/div[@class='profileCard profileCard--practitioner' and 1]/div[@class='profileCard__content' and 1]/div[@class='profileCard__name' and 2]/h5[@class='profileCard__title' and 1]/a[1]"
        elem = driver.find_element_by_xpath(person_id).click()
        time.sleep(2)
        
        div2 = ""
        div3 = ""
        div4 = ""
        div5 = ""
        div2link = ""
        div3link = ""
        div4link = ""
        div5link = ""
        
        p1 = ""
        p2 = ""
        p3 = ""
        list1 = ""
        list2 = ""
        list3 = ""

        name = driver.find_element_by_xpath("//h1[@class='profileHeader__title']")
        firstname = name.text.partition(" ")[0]
        restofname = name.text.partition(" ")[2]
        lastname = restofname.partition(", ")[0]

        while True:
            try:
                div2 = driver.find_element_by_xpath("//div/div[@class='contactInfo']/div[@class='contactInfo__item'][2]").text
                div2link = driver.find_element_by_xpath("//div/div[@class='contactInfo']/div[@class='contactInfo__item'][2]/a[@class='_blank']").get_attribute("href")
                break
            except:
                break

        while True:
            try:
                div3 = driver.find_element_by_xpath("//div/div[@class='contactInfo']/div[@class='contactInfo__item'][3]").text
                div3link = driver.find_element_by_xpath("//div/div[@class='contactInfo']/div[@class='contactInfo__item'][3]/a[@class='_blank']").get_attribute("href")
                break
            except:
                break

        while True:
            try:
                div4 = driver.find_element_by_xpath("//div/div[@class='contactInfo']/div[@class='contactInfo__item'][4]").text
                div4link = driver.find_element_by_xpath("//div/div[@class='contactInfo']/div[@class='contactInfo__item'][4]/a[@class='_blank']").get_attribute("href")
                break
            except:
                break

        while True:
            try:
                div5 = driver.find_element_by_xpath("//div/div[@class='contactInfo']/div[@class='contactInfo__item'][5]").text
                div5link = driver.find_element_by_xpath("//div/div[@class='contactInfo']/div[@class='contactInfo__item'][5]/a[@class='_blank']").get_attribute("href")
                break
            except:
                break
                

        while True:
            try:
                p1 = driver.find_element_by_xpath("//div[@class='pageSubSection pageSubSection--text']/p[1]").text
                p2 = driver.find_element_by_xpath("//div[@class='pageSubSection pageSubSection--text']/p[2]").text
                p3 = driver.find_element_by_xpath("//div[@class='pageSubSection pageSubSection--text']/p[3]").text
                break
            except:
                break

        while True:
            try:
                list1 = driver.find_element_by_xpath("//div[@class='pageSubSection pageSubSection--text']/ul[1]/li").text
                list2 = driver.find_element_by_xpath("//div[@class='pageSubSection pageSubSection--text']/ul[2]/li").text
                list3 = driver.find_element_by_xpath("//div[@class='pageSubSection pageSubSection--text']/ul[3]/li").text
                break
            except:
                break
        time.sleep(1)
        df = pd.DataFrame(np.array([[firstname,lastname,state,div2,div2link,div3,div3link,div4,div4link,div5,div5link,p1,p2,p3,list1,list2,list3]]), columns=['FirstName','LastName', 'state','div2', 'div2link','div3','div3link','div4','div4link','div5','div5link','p1','p2','p3','list1','list2','list3']).append(df, sort=False)
        time.sleep(1)
        driver.execute_script("window.history.go(-1)")
        time.sleep(1)
    driver.find_element_by_xpath("//a[@class='next page-numbers']").click()  
```
