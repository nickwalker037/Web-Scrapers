```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.select import Select
import time
```

```python
options = webdriver.ChromeOptions() 
options.add_argument("download.default_directory=C:/Users/nicholas/Desktop/Python/Chatbot/Untitled Folder/DataDownload")
```

```python
driver = webdriver.Chrome(r'C:\Users\nicholas\Downloads\chromedriver_win32(1)\chromedriver.exe', chrome_options=options)  # Optional argument, if not specified will search path.
driver.get("http://nhfpl.org/collections/business-nonprofits-and-jobs/")

#content = driver.find_element_by_name('content')
elem = driver.find_element_by_xpath('//a[@href="'+'http://www.referenceusa.com/Static/Home'+'"]')
window_before = driver.window_handles[0]
driver.implicitly_wait(3)
elem.click()
window_after = driver.window_handles[1]
driver.switch_to_window(window_after)

driver.find_element_by_xpath("//*[@id='chkAgree']").click()
driver.implicitly_wait(3)
driver.find_element_by_xpath("//*[@class='originButton ui-priority-primary action-agree tcButton']").click()

driver.implicitly_wait(3)
driver.find_element_by_xpath("//*[@id='matchcode']").send_keys("25000023426803")
driver.find_element_by_xpath("//*[@class='originButton ui-priority-primary default-action action-submit-form']").click()

driver.implicitly_wait(3)
driver.find_element_by_xpath('//a[@href="'+'/UsBusiness/Search'+'"]').click()
driver.find_element_by_xpath("//*[@class='advancedSearch']").click() # click on 'Advanced Search' tab


###################################################################
# Searching for specific fields within the 'Advanced Search' option
###################################################################

# MetroArea Search Field
driver.find_element_by_xpath("//*[@id='cs-MetroArea']").click() # click 'Metro Area' on left menu
driver.implicitly_wait(3)

s1 = Select(driver.find_element_by_xpath("//select[@id='usMsaState']")) # Select the dropdown menu
s1.select_by_value("DC") # Select 'DC' from the dropdown menu
driver.implicitly_wait(3)

driver.find_element_by_xpath("//*[@id='availableMSA']/ul/li/span").click() # click the menu option

driver.find_element_by_xpath("//a[@class='standardButtons blueMedium action-update-count default-action']").click() #update count


# 'Major Industry Group' Search Field
driver.find_element_by_xpath("//*[@id='cs-MajorIndustryGroup']").click()
driver.find_element_by_xpath("//*[@id='migTreeView']/ul/li[9]/div").click() # expand 'Services' tree
driver.find_element_by_xpath("//*[@id='migTreeView']/ul/li[9]/ul/li[2]/div").click() #expand 'Personal Services' tree
driver.find_element_by_xpath("//*[@id='migTreeView']/ul/li[9]/ul/li[2]/ul/li[10]/label").click() #select 'Beauty Shops'
driver.find_element_by_xpath("//*[@id='migTreeView']/ul/li[9]/ul/li[2]/ul/li[11]/label").click() #select 'Barber Shops'

driver.implicitly_wait(3)
driver.find_element_by_xpath("//a[@class='standardButtons blueMedium action-update-count default-action']").click() #update count


# 'Web Address' Search Field
driver.find_element_by_xpath("//*[@id='cs-WebAddress']").click() # check 'Web Address' on left menu
driver.find_element_by_xpath("//*[@id='phWebAddress']/div/div[2]/div/fieldset/ol/li/input").click() # check only companies with a web address






#################################
# Viewing Results of Search Query
#################################
driver.find_element_by_xpath("//a[@class='standardButtons greenMedium action-view-results']").click() #View Results
```

```python
max_page = driver.find_element_by_xpath("//*[@class='data-page-max']").text
current_page = driver.find_element_by_xpath("//*[@class='data-page-number']").text

max_page = int(max_page)
current_page = int(current_page)

while current_page < max_page - 3:
    if current_page > 1:
        #####################################
        # Unchecking pages from last download
        #####################################

        driver.find_element_by_xpath("//*[@id='checkboxCol']").click() # uncheck results
        driver.find_element_by_xpath("//*[@class='prev button mousedown-enterkey']").click() # previous page
        time.sleep(2)
        driver.find_element_by_xpath("//*[@id='checkboxCol']").click() # uncheck results
        driver.find_element_by_xpath("//*[@class='prev button mousedown-enterkey']").click() # previous page
        time.sleep(2)
        driver.find_element_by_xpath("//*[@id='checkboxCol']").click() # uncheck results

        driver.find_element_by_xpath("//*[@class='next button mousedown-enterkey']").click() # next page
        driver.find_element_by_xpath("//*[@class='next button mousedown-enterkey']").click() # next page
        driver.find_element_by_xpath("//*[@class='next button mousedown-enterkey']").click() # next page
    
    
    #####################################
    # Downloading Results of Search Query
    #####################################

    # Page 1:
    time.sleep(2)
    driver.implicitly_wait(3)
    driver.find_element_by_xpath("//*[@id='checkboxCol']").click() # Check all on page
    driver.find_element_by_xpath("//*[@class='next button mousedown-enterkey']").click() # next page
    # Page 2:
    time.sleep(2)
    driver.implicitly_wait(3)
    driver.find_element_by_xpath("//*[@id='checkboxCol']").click() # Check all on page
    driver.find_element_by_xpath("//*[@class='next button mousedown-enterkey']").click() # next page
    # Page 3:
    time.sleep(2)
    driver.implicitly_wait(3)
    driver.find_element_by_xpath("//*[@id='checkboxCol']").click() # Check all on page

    # Download: 
    driver.implicitly_wait(3)
    driver.find_element_by_xpath("//*[@class='action download action-download']").click() # Go to download page
    driver.implicitly_wait(3)
    driver.find_element_by_xpath("//*[@id='detailDetail']").click() # select detailed export
    driver.implicitly_wait(3)
    driver.find_element_by_xpath("//*[@class='originButton ui-priority-primary view-on-exportlimit view-on-randomsample action-download']").click() # download records
    driver.implicitly_wait(3)
    
    ############################
    # Go Back to List of Results
    ############################
    driver.find_element_by_xpath("//*[@class='backButton action-back']").click() # go back
    
    
    ################
    # Increase step
    ################
    time.sleep(2)
    current_page = driver.find_element_by_xpath("//*[@class='data-page-number']").text
    current_page = int(current_page)
    print(current_page)
```
